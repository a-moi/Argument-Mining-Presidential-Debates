```python
import csv
```

```python
sents = []
with open('updated_csv.csv', encoding="UTF-8") as file:
    read = csv.reader(file)
    for i in read:
        sents.append(i[1])
```

```python
from nltk.lm.preprocessing import pad_both_ends
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
def bigrams(corpus):
    padded = []
    final_lst = []
    token = []
    for i in corpus:
        if i != 'Sentences':
            token.append(word_tokenize(i))
    #print(token)
    for j in token:
        padding_process = list(pad_both_ends(j, 2))
        for i in padding_process:
            padded.append(i)
    ngram = ngrams(padded, 2)
    # iterate through the n-gram generator and append them to final_lst in order to return all n-grams
    # otherwise, the code would return only one n-gram
    for i in ngram:
        final_lst.append(i)
    return final_lst
bigrams(sents)
```

```python
from nltk.lm.preprocessing import pad_both_ends
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
def trigrams(corpus):
    padded = []
    final_lst = []
    token = []
    for i in corpus:
        # removing the header
        if i != 'Sentences':
            token.append(word_tokenize(i))
    for j in token:
        padding_process = list(pad_both_ends(j, 2))
        for i in padding_process:
            padded.append(i)
    ngram = ngrams(padded, 3)
    for i in ngram:
        final_lst.append(i)
    return final_lst
trigrams(sents)
```
